# Boosted portfolios

Choose top 2 stocks based on signal from boosted.ai and trade with **50**% allocation for each position


```python
import pandas as pd

symbol = 'BRX'
from_date = '2023-07-31'
to_date = '2023-08-24'
```

```python
token = ''
```

If you're looking to access a more extensive range of data, please visit https://datahub.limex.com to obtain your API token. With this token, you'll unlock the full potential of our datasets and can seamlessly integrate them into your projects

```python
boostedsignals = pd.read_json('https://hub.limex.com/v1/signals?vendor=boosted&token='+token+'&model=50678d2d-fd0f-4841-aaee-7feac83cb3a1&from=2022-01-01')
boostedsignals
```

```python
signals = boostedsignals.groupby('trade_date').apply(lambda x: x.nlargest(2, 'allocation')).reset_index(drop=True)
```

```python
universe = signals['symbol'].unique()
universe
```

```python
# Function to download stock data for a symbol
def download_stock_data(symbol):
  try:
    stock_data = pd.read_json('https://hub.limex.com/v1/candles?symbol='+symbol+'&token='+token+'&from=2022-01-01')
    stock_data['symbol'] = symbol  # Add a 'Symbol' column to identify the stock
    return stock_data
  except:
    print(symbol+' not availible')

# Download stock data for each symbol and store in a list
dataframes = [download_stock_data(symbol) for symbol in universe]

# Concatenate the list of DataFrames into one
stock_df = pd.concat(dataframes)
```

```python
stock_df['Future_Return'] = stock_df.groupby('symbol')['c'].pct_change(1) * 100
stock_df
```
